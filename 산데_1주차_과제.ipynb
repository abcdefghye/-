```python
import os
data_path="/content/drive/MyDrive/산업데이터과학"
data_name=os.listdir(data_path)
iris0=pd.read_csv(data_path + '/' + data_name[0])
iris0.head()
```

```python
iris=iris0.loc[:,['Species','SepalLengthCm']]
iris.head()
```

```python
iris1=iris[iris['Species']=='Iris-setosa']
iris2=iris[iris['Species']=='Iris-versicolor']
iris3=iris[iris['Species']=='Iris-virginica']
```

```python
iris11=iris1['SepalLengthCm'].sum()
iris22=iris2['SepalLengthCm'].sum()
iris33=iris3['SepalLengthCm'].sum()
```

```python
iris4=pd.DataFrame({"Species":['Iris-setosa'],"SepalLengthCm":[iris11]})
iris5=pd.DataFrame({"Species":['Iris-versicolor'],"SepalLengthCm":[iris22]})
iris6=pd.DataFrame({"Species":['Iris-virginica'],"SepalLengthCm":[iris33]})
```

```python
merge_iris=pd.concat([iris4, iris5, iris6], axis=0)
merge_iris.set_index('Species')
merge_iris.head()
```
